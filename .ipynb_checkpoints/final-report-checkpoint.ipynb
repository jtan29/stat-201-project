# Alcohol Consumption and Student Performance

## Introduction

Drinking alcohol is a common aspect of daily life for many individuals. Additionally, numerous students resort to alcohol consumption as a means of coping with stress. However, alcohol can have detrimental effects on people's bodies and nerves(1), leading to aggressive behavior (2). Our group aimed to explore whether alcohol consumption influences the academic performance of different types of students. According to a study conducted by Ansari, Stock, and Mills in 2013, alcohol consumption is inversely associated with student achievement (3).


For our group project, we opted to utilize a dataset named "Student Alcohol Consumption", this dataset is available at (https://www.kaggle.com/datasets/uciml/student-alcohol-consumption). This dataset includes information on students' gender, age, length of study, number of absences, weekend alcohol consumption (rated on a scale from 1 to 5, with 1 indicating very low and 5 indicating very high), and final grades (G3). The central question we aim to investigate is whether weekend drinking has a more pronounced effect on students' academic grades. We plan to arrive at our final conclusion by calculating a location parameter, the difference in mean student grade between students with high alcohol consumption and students with low alcohol consumption. We will compare the mean scores of both groups in an attempt to draw a conclusion regarding the impact of high alcohol consumption on student performance. Additionally, we are interested in investigating the standard deviation of test scores in both groups as a scale parameter.

## Method and Results

### Preliminary Analysis and Results

#### Preparing Data

We begin by preparing the required packages and reading our data set from the web.

```r
library(tidyverse)
library(tidymodels)
library(repr)
options(repr.matrix.max.rows = 6)
library(digest)
library(infer)
library(repr)
```

```r
url <-"https://raw.githubusercontent.com/jtan29/stat-201-project/main/student-mat.csv"
student_data <- read_csv(url)
head(student_data)
```

In our investigation of the effects of high alcohol consumption on student performance, we divide students into two groups: students reporting high alcohol consumption and students reporting low alcohol consumption. The students' alcohol consumption was assessed on an index from 1 to 5, with 1 being very low and 5 being very high. We have designated students with an index of 3 or higher as belonging to the high-consumption group and students with an index of 2 or below as belonging to the low-consumption group. The following code produces another column in the data set corresponding to the group the student belongs to.

```r
student_data_clean <- 
    student_data %>%
    select(Walc, G3) %>%
    mutate(high_alc = Walc > 2)

head(student_data_clean)
```

#### Analysis of Raw Data

We first determine the mean final grade and the standard deviation of grades for the high and low consumption groups.

```r
student_summary <-
    student_data_clean %>%
    group_by(high_alc) %>%
    summarize(count = n(),
             mean_grade = mean(G3),
             sd_grade = sd(G3))

low_alc_mean = student_summary$mean_grade[[1]]
high_alc_mean = student_summary$mean_grade[[2]]
total_count = student_summary$count[[1]] + student_summary$count[[2]]
student_summary
total_count
```

Then we plot the sample distribution of student grades for both groups.

```r
options(repr.plot.width = 12, repr.plot.height = 8)

high_alc_sample_dist <-
    student_data_clean %>%
    filter(high_alc == TRUE) %>%
    ggplot(aes(x = G3)) +
    geom_histogram(binwidth = 1, fill = "lightblue",  color  = "blue") +
    labs(title = "Sample distribution of students with high alcohol consumption", x = "Final grade (out of 20)") +
    geom_vline(aes(xintercept = high_alc_mean), linetype = "dashed", color = "red", linewidth = 1) +
    coord_cartesian(xlim = c(0, 20))


high_alc_sample_dist
    
low_alc_sample_dist <-
    student_data_clean %>%
    filter(high_alc == FALSE) %>%
    ggplot(aes(x = G3)) +
    geom_histogram(binwidth = 1, fill = "lightblue",  color  = "blue") +
    labs(title = "Sample distribution of students with low alcohol consumption", x = "Final grade (out of 20)") +
    geom_vline(aes(xintercept = low_alc_mean), linetype = "dashed", color = "red", linewidth = 1) +
    coord_cartesian(xlim = c(0, 20))


low_alc_sample_dist
```

We can also visually inspect a boxplot of the two groups.

```r
boxplot <- student_data_clean %>%
              ggplot() +
              geom_boxplot(aes(x = high_alc, y = G3, fill = high_alc)) +
              labs(x = "High alcohol consumption reported", 
                   y = "Test score", 
                   fill = "High alcohol consumption reported", 
                   title = "Comparison of test scores among students with high/low alcohol consumption")
boxplot
```

Our anticipated result, based on existing literature on alcohol consumption, is that students with lower alcohol consumption receive higher grades on average (alternatively, this can be restated as students with higher reported alcohol consumption receiving lower grades on average). This is supported by the lower raw mean final grade in the high consumption group in our preliminary results, as can be observed in the boxplot above. However, the difference is small and could be attributed to random chance in the sampling process. To better determine if this difference is truly significant, we turn to hypothesis testing.

### Hypothesis Testing

First, we must establish our null and alternative hypotheses. As stated in the preceding subsection, our initial assumption is that students with low alcohol consumption would achieve higher G3 grades compared to students with high alcohol consumption. This assumption forms the alternative hypothesis in this analysis, with the null hypothesis being that students consuming high volumes of alcohol achieve the same grades as students with low alcohol consumption. We can express the two hypotheses as follows:

$$H_0: \mu_1 - \mu_2 = 0$$ 
$$H_a: \mu_1 - \mu_2 > 0$$

Where $\mu_1$ is the mean test score of all students with low alcohol consumption, and $\mu_2$ is the mean test score of all students with high alcohol consumption

As part of hypothesis testing, we will calculate a p-value to determine whether to reject our null hypothesis. We set our significance level at $\alpha = 0.05$. However, just reporting the p-value does not capture all the relevant information about the two groups and the relationship between alcohol and student performance. We will also calculate a 95% confidence interval to provide an estimated interval which is likely to capture our unknown population parameter (the difference in mean student grades). Additionally, we intend to find the p-value and confidence interval using both non-traditional computer-based models and traditional theoretical models such as the t-distribution.

#### Simulation-Based Hypothesis Tests

We will begin our analysis with a simulation-based hypothesis test using the `infer` package in R. As a first step, we create the null model for our data.

```r
set.seed(39)
null_sim <- student_data_clean  %>%
  specify(G3 ~ high_alc) %>%
  hypothesize(null = "independence") %>%
  generate(reps = 2000, type = "permute") %>%
  calculate(
    stat = "diff in means",
    order = c("FALSE", "TRUE")
  )

null_sim %>% visualize()
```

Next, we will calculate the difference between the mean student score in each sample, which is our point statistic:

```r
obs_diff <- student_data_clean %>%
  specify(G3 ~ high_alc) %>%
  calculate(
    stat = "diff in means",
    order = c(FALSE, TRUE)
  )
obs_diff
```

Below, we present the simulation-based null-model with the observed point statistic shaded in red.

```r
null_sim %>%
  visualize() +
  shade_p_value(obs_stat = obs_diff, direction = "greater")
```

Which leads us to the simulation-derived p-value of:

```r
pvalue <- null_sim %>%
  get_pvalue(obs_stat = obs_diff, direction = "greater")
pvalue
```

Our simulation-based hypothesis test finds a p-value of $0.325$.

We can then find 95% confidence interval for the difference of means, again with the `infer` package.

```r
boot_ci <- student_data_clean %>%
  specify(G3 ~ high_alc) %>%
  generate(reps = 2000, type = "bootstrap") %>%
  calculate(
    stat = "diff in means",
    order = c(FALSE, TRUE)
)

conf_in <-   boot_ci %>%
    get_ci(level = 0.95)

conf_in

boot_ci %>% visualize() +
    shade_ci(endpoints = conf_in)
```

Our 95% confidence interval provides us with a range of values where we would expect the true population parameter to land. Our bootstrap-based confidence interval places these bounds as a difference in mean score of $-0.701$ (students with low alcohol consumption score $0.701$ points below students with high alcohol consumption) and $1.093$ (students with low alcohol consumption score $1.093$ higher than students with high alcohol consumption). It is important here to note that the value of $0$ (students in either group performed the same) is included in this interval.

#### Theory-Based Hypothesis Testing

After obtaining a value from our simulation-based hypothesis testing, we are interested in determining any potential differences between our simulation-based test and a theory-based test. From our preliminary analysis, our data does not fully follow a normal distribution. However, the sample size is large enough for the Central Limit Theorem to apply, and we determine that it is suitable to use traditional theoretical hypothesis testing methods on the data. In particular, the most suitable test is a t-test for a difference in means of two independent groups, as the population's standard deviation is not known.

$$t =\frac{ (\bar{x}_1 - \bar{x}_2)}{ \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}} \sim t (df = n_1 + n_2 - 2)$$

```r
student_summary
```

Our preliminary summary information provides us with all the information we need to calculate the p-value using a t-distribution with $n_1 + n_2 - 2 = 236 + 159 - 2 = 393$ degrees of freedom.

```r
high_alc_mean <- student_summary$mean_grade[[1]]
low_alc_mean <- student_summary$mean_grade[[2]]
high_alc_sd <- student_summary$sd_grade[[1]]
low_alc_sd <- student_summary$sd_grade[[2]]
high_alc_n <- student_summary$count[[1]]
low_alc_n <- student_summary$count[[2]]

test_statistic <- abs(high_alc_mean - low_alc_mean) / sqrt(high_alc_sd ** 2 / high_alc_n + low_alc_sd ** 2 / low_alc_n)
p_value <- pt(test_statistic, df = high_alc_n + low_alc_n - 2, lower.tail = FALSE)
p_value
```

Our theory-based hypothesis test finds a p-value of $0.318$.

We can then complete this analysis by providing a theory-based confidence interval for the difference in means.

```r
upper_ci <- abs(high_alc_mean - low_alc_mean) + qt(0.975, df = high_alc_n + low_alc_n - 2) * sqrt(high_alc_sd ** 2 / high_alc_n + low_alc_sd ** 2 / low_alc_n)
lower_ci <- abs(high_alc_mean - low_alc_mean) - qt(0.975, df = high_alc_n + low_alc_n - 2) * sqrt(high_alc_sd ** 2 / high_alc_n + low_alc_sd ** 2 / low_alc_n)
upper_ci
lower_ci
```

We find a theory-based confidence interval spanning from $\approx -0.663$ to $\approx 1.085$. This is consistent with our bootstrap distribution, and also includes zero within the interval. We can conclude that for our hypothesis test, the computer-based models produced similar results to the traditional theoretical models. This is not unexpected, as we previously assessed whether our data fulfilled the requirements for theoretical models such as the t-distribution to be effective. The similarity to the computer-based results, which have less stringent requirements on the data set, supports our assessment that traditional theoretical techniques were suitable in this case.

### Confidence Interval for the Standard Deviation

We may also be interested in observing the standard deviation among test scores of students in both groups. Below, we present the bootstrap confidence intervals for the standard deviations of our high and low alcohol groups.

```r
boot_low_sd <- student_data_clean %>%
  filter(high_alc == FALSE) %>%
  specify(response = G3) %>%
  generate(reps = 2000, type = "bootstrap") %>%
  calculate(
    stat = "sd"
)

conf_in_low <-   boot_low_sd %>%
    get_ci(level = 0.95)

conf_in_low

boot_low_sd %>% visualize() +
    shade_ci(endpoints = conf_in_low) +
    ggtitle("Standard deviation of test scores (low alcohol) Bootstrap Distribution")
```

```r
boot_high_sd <- student_data_clean %>%
  filter(high_alc == TRUE) %>%
  specify(response = G3) %>%
  generate(reps = 2000, type = "bootstrap") %>%
  calculate(
    stat = "sd"
)

conf_in_high <-   boot_high_sd %>%
    get_ci(level = 0.95)

conf_in_high

boot_high_sd %>% visualize() +
    shade_ci(endpoints = conf_in_high) +
    ggtitle("Standard deviation of test scores (high alcohol) Bootstrap Distribution")
```

This would suggest that is much greater variability in the test scores of students that did not consume high volumes of alcohol.

# Discussion

Our hypothesis tests, both computer-based and theoretical, found similar p-values of $0.325$ and $0.318$ respectively. Both these values are much greater than our $\alpha = 0.05$. As a result, we **fail to reject the null hypothesis** and find that there is not sufficient evidence to support our alternative hypothesis that students who consume lower volumes of alcohol will have higher test scores. This is a somewhat unexpected finding, as our initial expectations were for students who consumed less alcohol would achieve high test scores compared to students that consumed more alcohol. We believe that some of this may be attributed to the large number of students in the lower alcohol group reporting a final grade of 0, which lowers the mean of the entire group. In the sample distributions of both groups, it can be seen that a significant number of students received a grade of zero. In particular, 30 of the 236 students in the low alcohol group reported a score of zero. This may indicate a problem with the data collection, particularly with the handling of students that did not report a test score or did not follow through with the entire data collection process.

Also of concern is how the alcohol levels are determined, since there is no direct numerical measurement of the actual alcohol consumed. It is possible that students that consume the same amount of alcohol will self-assess into different groups based on their perception of what constitutes high and low alcohol consumption. Additionally, if the alcohol levels are self-reported, students may not feel comfortable reporting their actual consumption habits.

Future investigations could focus on revising how the data is collected to prevent students presenting invalid scores from being included in the data set. Future questions this could lead to may include investigations seek to more finely quantify the volume of alcohol consumed, in order to investigate the volume needed for changes in grades to be observed.
